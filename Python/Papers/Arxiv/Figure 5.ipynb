```python
# Python Code for Figure 5 in `CV QKD with single quadrature measurement at arbitrary reference frame'

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

colnames = ['a','theta']
df_1 = pd.read_csv("DataBob20-1000-2.txt", names=colnames, delimiter=' ', header=None, dtype=float)
df_2 = pd.read_csv("DataBob50-1000.txt", names=colnames, delimiter=' ', header=None, dtype=float)
df_1 = df_1.iloc[2900:8000]
print(df_1)
df_2 = df_2.iloc[:1500]
df_1 = df_1.reset_index()
df_2 = df_2.reset_index()
df_1['index'] = df_1['index'].div(1000).round(2)
df_1['index'] = df_1['index']-2.9
df_2['index'] = df_2['index'].div(1000).round(2)
df_1['theta']=np.pi*(df_1['theta']-df_1['theta'].min())/(df_1['theta'].max()-df_1['theta'].min())
df_2['theta']=np.pi*(df_2['theta']-df_2['theta'].min())/(df_2['theta'].max()-df_2['theta'].min())/2

df_1 = df_1.round({'theta': 3})
df_1 = df_1.replace(0.000, -1)
print(df_1)
df_list = [df_1 ,df_2]
fig, axes = plt.subplots(1, 2,figsize=(10,4))
fig1 = df_1.plot(kind='scatter', x='index', y='theta',color='black',s=1,ax=axes[0])
fig2 = df_2.plot(kind='scatter', x='index', y='theta',color='black',s=1,ax=axes[1])

fig2a = df_2.plot(kind='scatter', x='index', y='theta',color='black',s=1)
fig1a.set_xlabel("Data (/ms)")
fig1a.set_ylabel("phase (/rad)")
fig2a.set_xlabel("Data (/ms)")
fig2a.set_ylabel("phase (/rad)")
#plt.xlim([-0.2, 5.2])
#plt.ylim([-0.2, 3.3])
plt.savefig("Figure6b.pdf", bbox_inches='tight')

fig1.set_xlabel("Data (/ms)")
fig1.set_ylabel("phase (/rad)")
fig2.set_xlabel("Data (/ms)")
fig2.set_ylabel("phase (/rad)")
plt.show()

fig.savefig("Figure6.png", dpi = 2000)
fig.savefig("Figure6.pdf", bbox_inches='tight')
```
